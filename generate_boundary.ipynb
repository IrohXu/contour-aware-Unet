```python
import PIL.Image as Image
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
import math
```

```python
def gray(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_gray = np.dot(img_rgb[...,:3], [0.299, 0.587, 0.114])
    return img_gray
```

```python
img_path = 'u_net_liver/data/data/train/310_mask.png'
```

```python
path = 'u_net_liver/data/data/train'
```

```python
for filename in os.listdir(path): 
    tag = filename.split('.')[0][-1]
    if tag == 'k':
        image_path = os.path.join(path, filename)
        img = cv2.imread(image_path)
        img = gray(img)
        xgrad = cv2.Sobel(img, cv2.CV_16SC1, 1, 0) #x方向梯度
        ygrad = cv2.Sobel(img, cv2.CV_16SC1, 0, 1) #y方向梯度
        edge_output = cv2.Canny(xgrad, ygrad, 50, 150)
        edge = []
        for i in range(0, 512):
            for j in range(0,512):
                if edge_output[i][j] == 255:
                    edge.append((i,j))
        for element in edge:
            for i in range(-2, 3, 1):
                for j in range(-2, 3, 1):
                    try:
                        edge_output[element[0]+i][element[1]+j] = 255
                    except:
                        continue
        img_RGB = cv2.cvtColor(edge_output, cv2.COLOR_GRAY2BGR)
        save_path = os.path.join(path, filename[0:3] + '_contour.png')
        print(save_path)
        cv2.imwrite(save_path,img_RGB)
    else:
        continue
print("All finish")
```
